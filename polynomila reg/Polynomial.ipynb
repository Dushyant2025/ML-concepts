```python
What is Polynomial Regression?

Polynomial Regression is a type of linear regression — but it allows non-linear relationships between the input feature(s) and target variable.

Even though the curve looks non-linear, the model is still linear in parameters (the coefficients).

if Linear:
    y=Bo +B1x
if poly( degree=2):
    y=Bo+ B1x + B2x^2
if degree =3:
    y=Bo+ B1x + B2x^2 + B3x^3
    
```

```python
When to use:
| Situation                                                 | Why Polynomial Works                          |
| --------------------------------------------------------- | --------------------------------------------- |
| Data shows **U-shape** or **inverted U** trend            | Linear regression can’t capture curvature     |
| You notice **systematic error** in residuals (not random) | Indicates missing curvature                   |
| Relationship between variables **changes direction**      | Polynomial terms capture those changes        |
| Moderate dataset size                                     | Because higher-degree polynomials can overfit |
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
#Create a U-shaped (non-linear) dataset
np.random.seed(42)
X = np.linspace(-3, 3, 100).reshape(-1, 1)  # 100 points between -3 and 3
y = X**2 + np.random.randn(100, 1) * 2       # U-shape with noise (y = x^2 + noise)

# Plot the raw data
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='blue', label='Actual data (U-shaped)')
plt.title("Raw U-shaped Data (y = x² + noise)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# Trying LINEAR REGRESSION (it will fail)
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred_linear = lin_reg.predict(X)

plt.figure(figsize=(8,5))
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred_linear, color='red', label='Linear Regression Fit')
plt.title("Linear Regression on U-shaped Data (Poor Fit)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# Apply POLYNOMIAL REGRESSION (degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)   # Adds X^2 term

poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
y_pred_poly = poly_reg.predict(X_poly)

# Plot Polynomial Regression fit
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred_poly, color='red', label='Polynomial Regression (Degree 2)')
plt.title("Polynomial Regression Fit (Degree 2)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# Print coefficients
print("Linear Regression Coefficients:", lin_reg.coef_, "Intercept:", lin_reg.intercept_)
print("Polynomial Regression Coefficients:", poly_reg.coef_, "Intercept:", poly_reg.intercept_)
```
